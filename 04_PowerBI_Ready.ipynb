Create Power BI View

```python
spark.sql("""
CREATE OR REPLACE VIEW powerbi_customer_risk AS
SELECT
    SK_ID_CURR AS customer_id,
    AMT_INCOME_TOTAL AS income,
    AMT_CREDIT AS loan_amount,
    total_previous_loans,
    avg_previous_loan_amount,
    risk_score,
    risk_category,
    batch_date
FROM gold_customer_risk_profile
""")
```
